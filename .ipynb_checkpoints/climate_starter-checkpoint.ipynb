# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Look at contents of Measurement table
inspector = inspect(engine)
columns = inspector.get_columns('Measurement')
for column in columns:
    print(column['name'], column['type'])
```

```python
# Look at contents of Station table
columns = inspector.get_columns('Station')
for column in columns:
    print(column['name'], column['type'])
```

# Exploratory Climate Analysis

```python
# Calculate the date 1 year ago from the last data point in the database
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Perform a query to retrieve the data and precipitation scores
percip_last_year = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > '2016-08-23').all()
percip_last_year
```

```python
# Save the query results as a Pandas DataFrame
percip_df = pd.DataFrame(percip_last_year, columns=['date', 'prcp'])

# Dates are repeated due to multiple stations.  Get means for each date.
percip_avgs = percip_df.groupby('date').mean()

percip_avgs_df = pd.DataFrame(percip_avgs).reset_index(['date'])
percip_avgs_df.head()
```

```python
# Sort the dataframe by date
percip_avgs_df = percip_avgs_df.sort_index()
percip_avgs_df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
percip_avgs_df.plot.bar()

plt.title("Average Daily Precipitation Over 12 Months")
plt.xlabel("Date")
plt.ylabel("Precipitation")

plt.tight_layout()
plt.show()
```

# Temperature Analysis I

```python
#Identify the average temperature in June at all stations across all available years in the dataset
month_june = "06"
june_avg_temp = session.query(func.avg(Measurement.tobs)).\
    filter(func.strftime("%m", Measurement.date) == month_june, Measurement.date < "2016-12-31").\
    group_by(func.strftime("%Y", Measurement.date)).all()
june_avg_temp
```

```python
#Do the same for December temperature.
month_dec = "12"
dec_avg_temp = session.query(func.avg(Measurement.tobs)).\
    filter(func.strftime("%m", Measurement.date) == month_dec).\
    group_by(func.strftime("%Y", Measurement.date)).all()
dec_avg_temp
```

```python
#Use the t-test to determine whether the difference in the means, if any, is statistically significant.
stats.ttest_rel(june_avg_temp, dec_avg_temp)
#The p-value is <0.05 which means there is a significant difference between the temperatures in June and Dec.

#Will you use a paired t-test, or an unpaired t-test? Why?
#I used a paired t-test since the temperature data came from the same stations at different times.
#That means the data is related and must be paired.
```

# Temperature Analysis II

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

temp_min_avg_max = (calc_temps('2016-09-13', '2016-09-23'))
print(temp_min_avg_max)
```

```python
#Convert query into DataFrame
temp_df = pd.DataFrame(temp_min_avg_max, columns=['min_temp', 'avg_temp', 'max_temp'])
temp_df
```

```python
#Get average temperature
tavg = temp_df.loc[0,'avg_temp']
tavg
```

```python
#Get difference between max and min temperature
tdiff = temp_df.loc[0, "max_temp"] - temp_df.loc[0, "min_temp"]
tdiff
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
temp_df.plot.bar(y="avg_temp", yerr=tdiff)

plt.title("Trip Avg Temp")
plt.ylabel("Temperature")

plt.tight_layout()
plt.show()
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
